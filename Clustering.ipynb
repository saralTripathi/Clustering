1. What is unsupervised learning in the context of machine learning?
   -  Unsupervised Learning is a type of machine learning where the model is not provided with labeled data. Instead, the algorithm tries to find patterns, structures, or relationships in the input data on its own.
  -  Unsupervised learning helps uncover the hidden structure of data without the need for labeled examples, making it especially useful when labels are hard or expensive to obtain.
2. How does K-Means clustering algorithm work?
   - The K-Means clustering algorithm is a popular unsupervised learning method used to group data points into K clusters based on their similarity.
   - Step-by-Step Process of K-Means:
       - Decide how many clusters you want to form.
       - Randomly pick K points from the data as the initial "centroids" (center of each cluster).
       - For each data point, compute the distance to all centroids, and assign it to the nearest one.
       - Recalculate each centroid as the mean (average) of all the data points assigned to it.
       - Repeat steps 3 & 4 .
3. Explain the concept of a dendrogram in hierarchical clustering.
   - A dendrogram is a tree-like diagram used to visualize the results of hierarchical clustering. It shows how individual data points are merged step by step into clusters, and helps you decide how many clusters to choose.
   - A dendrogram is a visual summary of how clusters form step-by-step in hierarchical clustering, and it helps you choose the number of clusters by showing merge distances between points or groups.
4. What is the main difference between K-Means and Hierarchical Clustering.
   - K-Means:
      - Partition-based (divides data into K fixed clusters).
      - you must specify K before running the algorithm.
      - Produces flat clusters.
      - Faster and more efficient on large datasets.
      - Can produce different results on different runs (due to random initialization).
      - Assumes clusters are spherical and of similar size.
      - Sensitive to initial centroid positions and outliers.
  - Hierarchical:
      - Builds a tree of clusters (dendrogram) either bottom-up or top-down.
      - No need to specify K initially; you can decide later by cutting the dendrogram.
      - Produces a dendrogram (tree structure of clustering steps).
      - Slower and not suitable for very large datasets.
      - Always gives the same result (deterministic process).
      - Sensitive to noise and outliers as early merges are irreversible.
5. What are the advantages of DBSCAN over K-Means?
   - DBSCAN Automatically finds the number of clusters.
   - DBSCAN Can find clusters of any shape (e.g., curved, elongated).
   - DBSCAN Can identify outliers as noise (doesn't force every point into a cluster).
   - DBSCAN Same result every time you run it (no random initialization).
   - DBSCAN Starts with any unvisited point and expands based on density.
   - DBSCAN Can handle clusters of different densities, depending on parameter tuning.
6. When would you use Silhouette Score in clustering?
   - Use after clustering to check quality.
   - Helps determine the right number of clusters (especially in K-Means).
   - Useful for comparing different algorithms or parameters.
   -  Avoid if clusters are non-convex or uneven in size — score may be misleading in such cases.
7. What are the limitations of Hierarchical Clustering?
   - Can become computationally expensive as the number of data points increases.
   - A single outlier can affect early merges, leading to poor clustering results.
   - Unlike K-Means (which minimizes within-cluster variance), there’s no clear optimization objective.
   -  Requires a Distance Metric
   - There’s no fixed K; you must "cut" the dendrogram, which can be subjective.
   - Struggles with clusters of very different sizes and densities.
8. Why is feature scaling important in clustering algorithms like K-Means?
   - Feature scaling is very important in clustering algorithms like K-Means because it ensures that all features contribute equally to the distance calculations.
   - K-Means groups data based on how close points are to centroids.Distance is affected by the scale (unit) of features.
   - If one feature has a large range (e.g., income in thousands) and another has a small range (e.g., age in years), the large one will dominate the distance metric.
   - Scaling puts all features on the same scale.
   - Without scaling, clusters may be biased or incorrectly shaped.
   - If height (0–2m) and income (0–1,000,000) are used without scaling, K-Means will cluster based mostly on income, ignoring height.
9. How does DBSCAN identify noise points?
   - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points as those that do not belong to any cluster. It does this using two parameters: ε (epsilon) and MinPts (minimum points).
   - ε (epsilon):The radius around a data point to search for neighbors.
   - MinPts:The minimum number of points required to form a dense region (cluster).
   - DBSCAN checks each point:
       - If it has enough neighbors (MinPts within ε) → it's a core point.
       - If not, but it's close to a core point → it's a border point.
       - If it’s neither a core nor a border point → it is labeled as noise.
10.  Define inertia in the context of K-Means.
     - Inertia is a measure of how tightly the data points in a cluster are grouped around the centroids. It is also known as the within-cluster sum of squares (WCSS).
     - Measures the compactness of clusters.
     - Lower inertia implies tighter clusters.
     - Used in the Elbow Method to choose the best K.
     - Decreases with more clusters, but beware of overfitting.
11.  What is the elbow method in K-Means clustering?
     - The Elbow Method is a technique used to determine the optimal number of clusters (K) in K-Means by analyzing how inertia (within-cluster sum of squares) changes as K increases.
     - How It Works (Step-by-Step):
         - Run K-Means for a range of cluster numbers (e.g., K = 1 to 10).
         - For each K, compute the inertia (sum of squared distances of points to their nearest cluster center).
         - Plot the number of clusters (K) on the x-axis and the inertia on the y-axis.
         - Look for a point where the decrease in inertia slows down sharply — this point is called the "elbow."
         - The value of K at the elbow is considered the optimal number of clusters.
12. Describe the concept of "density" in DBSCAN.
    - In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the concept of density is the foundation for forming clusters.
    - Density refers to how closely packed the data points are in a region.
    - A region is considered dense if a minimum number of points (MinPts) are found within a specified radius (ε, epsilon) of a point.
    - DBSCAN uses ε and MinPts to define density.
13. Can hierarchical clustering be used on categorical data?
    - Yes, Hierarchical Clustering can be used on categorical data, but with some important adjustments.
    - Since hierarchical clustering is based on distances, and categorical data doesn't have natural numeric distances, you need to Use an Appropriate Distance Metric.
    - For categorical data, you cannot use Euclidean distance. Instead, use:
         - Hamming distance
         - Jaccard distance
         - Simple matching coefficient
         - Gower distance
14. What does a negative Silhouette Score indicate?
    - A negative silhouette score means that a data point is likely assigned to the wrong cluster.
15. Explain the term "linkage criteria" in hierarchical clustering.
    - In hierarchical clustering, the linkage criteria define how the distance between two clusters is calculated when forming the hierarchy (i.e., deciding which clusters to merge next).
    - Since hierarchical clustering builds clusters by iteratively merging the closest pair of clusters, the way you measure "closeness" between clusters affects the shape and size of the resulting clusters.
16.  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
     - K-Means performs poorly on data with varying cluster sizes/densities because it assumes equal-sized, spherical clusters with uniform density and uses Euclidean distance, which can misrepresent the real structure.
     - K-Means clustering assumes that clusters are spherical, equally sized, and equally dense. When the data violates these assumptions, K-Means can struggle to form meaningful clusters.
     - K-Means tries to split data into clusters of similar volume. It may misclassify smaller clusters or merge unequal-sized clusters.
     - Outliers can shift centroids, leading to poor clustering.
17.  What are the core parameters in DBSCAN, and how do they influence clustering.
     - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) relies on two main parameters:
     - 1. ε (epsilon) — Neighborhood Radius:
          - Defines the maximum distance between two points to be considered neighbors.
          - Influences how large or tight a cluster can be.
          - Too small ε → Many points are labeled as noise or form tiny clusters.
          - Too large ε → Distant points may be incorrectly grouped into the same cluster.
    - 2.MinPts (Minimum Points):
          - The minimum number of points required within an ε-radius to consider a point a core point (i.e., dense region).
          - Helps determine how dense a region must be to qualify as a cluster.
          - Too small MinPts → Can detect noise as small clusters.
          - Too large MinPts → May miss valid small clusters; more points will be labeled as noise.
18. How does K-Means++ improve upon standard K-Means initialization.
    - K-Means++ is a smarter way to initialize the centroids in K-Means clustering. It significantly improves accuracy, convergence speed, and stability of the algorithm.
    -  K-Means++ Initialization — Step-by-Step:
         - Randomly choose the first centroid from the data points.
         - For each remaining data point, compute its distance (D(x)) to the nearest chosen centroid.
         - Select the next centroid with probability proportional to D(x)^2.
         - Repeat step 3 until K centroids are chosen.
         - Proceed with the standard K-Means algorithm (assignment + update steps).

19. What is agglomerative clustering?
    - Agglomerative Clustering is a type of hierarchical clustering that builds clusters bottom-up, by successively merging the closest pairs of clusters until all data points are grouped into a single cluster (or a desired number of clusters is reached).
    - Start with each data point as its own cluster.
    - At each step, merge the two closest clusters based on a linkage criterion.
    - Repeat until:
        - All points are in one big cluster (to build a dendrogram), or
        - A specific number of clusters is formed.
20. What makes Silhouette Score a better metric than just inertia for model evaluation?
    - Inertia measures the within-cluster sum of squared distances to the centroid.Lower inertia → tighter clusters.It always decreases as the number of clusters increases.So, it can't tell you the optimal number of clusters on its own.Inertia only considers compactness, not separation.
    - Silhouette Score measures how well each point fits within its cluster versus others.Silhouette Score considers both cohesion and separation.Silhouette is normalized, so it's easier to compare across models.Silhouette helps identify the best number of clusters.





































```python
#1.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans


X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.5, random_state=0)


kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)


plt.scatter(X[:, 0], X[:, 1], c=y_kmeans,s = 50, cmap='viridis')

centers = kmeans.cluster_centers_
centers

plt.scatter(centers[:,0],centers[:,1],c='black')
plt.show()
```

```python
#4. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

data = load_wine()
x = data.data

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

kmeans = KMeans(n_clusters=3, random_state=0,n_init=10)
kmeans.fit(x_scaled)

cluster_sizes = {}
for label in kmeans.labels_:
    if label in cluster_sizes:
        cluster_sizes[label] += 1

    else:
        cluster_sizes[label] = 1

print("Cluster Sizes:")
for label, size in cluster_sizes.items():
    print(f"Cluster {label}: {size} points")
```

```python
#9.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


silhouette_scores = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(x)
    y_means = kmeans.fit_predict(x)
    silhouette_avg = silhouette_score(x, y_means)
    silhouette_scores.append(silhouette_avg)

plt.bar(range(2, 6), silhouette_scores)
plt.title('silhouette scores for k = 2 to 5')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
#11.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

x, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.5, random_state=0)

kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(x)


x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),np.arange(y_min, y_max, 0.02))

z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)

plt.figure(1, figsize=(8, 6))
plt.imshow(z, interpolation='nearest',extent=(xx.min(), xx.max(), yy.min(), yy.max()),cmap='viridis')

plt.scatter(x[:, 0], x[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('K-Means Clustering with Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#14. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_breast_cancer()
x = data.data

inertia_values = []
for k in range(2,7):
  kmeans = KMeans(n_clusters = k,random_state = 0, n_init=10) # Added n_init for reproducibility
  kmeans.fit(x)
  inertia = kmeans.inertia_
  inertia_values.append(inertia)
  print(f"For K = {k}, Inertia = {inertia}")

plt.plot(range(2,7),inertia_values,marker = 'o')
plt.title('Inertia Values for K = 2 to 6')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python
#23.Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side.
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans,DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

x,_ = make_circles(n_samples = 300,noise = 0.05,factor = 0.5,random_state = 0)

kmeans = KMeans(n_clusters = 2,random_state = 0,n_init=10)
kmeans.fit(x)

dbscan  = DBSCAN(eps = 0.3,min_samples = 5)
dbscan.fit(x)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(x[:, 0], x[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(x[:, 0], x[:, 1], c=dbscan.labels_, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

```python
#24.Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


data = load_iris()
x = data.data

silhouette_scores = []
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(x)
    silhouette_avg = silhouette_score(x, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

plt.figure(figsize = (12,8))
plt.plot(range(2, 7), silhouette_scores, marker='o')
plt.title('Silhouette Coefficient for Each Sample')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

```python
#25.Generate synthetic data using make_blobs and apply Agglomerative Clustering with 'average' linkage. Visualize clusters.
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

x,_ = make_blobs(n_samples = 300,centers = 3,cluster_std = 0.5,random_state = 0)

agglo = AgglomerativeClustering(n_clusters = 3,linkage = "average")
agglo.fit(x)

plt.scatter(x[:,0],x[:,1],c = agglo.labels_,cmap = 'viridis')
plt.title('Agglomerative Clustering with Average Linkage')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#26. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = load_wine()
x = data.data
wine_df = pd.DataFrame(data.data,columns = data.feature_names)


kmeans = KMeans(n_clusters = 3,random_state = 0, n_init = 10)
kmeans.fit(x)
wine_df['cluster'] = kmeans.labels_

sns.pairplot(wine_df, vars=wine_df.columns[:4], hue='cluster', palette='viridis')
plt.suptitle('K-Means Clustering on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

```python
#27.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

x,_ = make_blobs(n_samples = 300,centers = 3,cluster_std = 0.5,random_state = 0)

dbscan = DBSCAN(eps = 0.5,min_samples = 5)
dbscan.fit(x)

# Count clusters and noise points
labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Estimated number of clusters: {n_clusters}")
print(f"Estimated number of noise points: {n_noise}")
```

```python
#28.Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

data = load_digits()
x = data.data

tsne = TSNE(n_components=2)
x_reduced = tsne.fit_transform(x)

agglo = AgglomerativeClustering(n_clusters = 10)
agglo.fit(x_reduced)

plt.scatter(x_reduced[:,0],x_reduced[:,1],c = agglo.labels_,cmap = 'viridis')
plt.title('Agglomerative Clustering on Digits (t-SNE Reduced)')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.show()
```
